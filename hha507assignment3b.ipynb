# **Loading in Packages**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# **Seaborn Visualizations**

```python
# Getting a list of datasets available in seaborn
sns.get_dataset_names()
```

### **Pairplot**

> Using seaborn pairplot to visualize data.



```python
# Loading in the 'flights' datasets from seaborn
df = sns.load_dataset('flights')
df
```

```python
# The 'flights' dataset is visualized with seaborn pairplot.
## As shown in the graph, the number of passengers increases with each passing year.
## In addition, the pairplot also shows increased passenger counts during certain months, which indicates increased traveling during certain months.
sns.pairplot(df, hue='month', x_vars=['year'], y_vars=['passengers'], height=3, aspect=1.5)
```

### **Heatmap**

> Using seaborn heatmap to visualize data.



```python
# The seaborn 'flights' data is also used for the heatmap visualization

## Using pivot, the data is reorganized into a matrix in order to plot the heatmap.
## The 'month' column will become the index, the 'year' column would become the new columns, and 'passengers' will fill out the values of the cells.
pivot = df.pivot("month", "year", "passengers")
```

```python
# Creating a heatmap using seaborn and matplotlib.pyplot
plt.figure(figsize=(10, 7))

sns.heatmap(pivot, annot=True, fmt="d", cmap="YlGnBu", cbar=True)

plt.title("Number of Passengers by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()

## As shown on the heatmap, the number of passengers started to increase around 1952/1953.
## There is a larger amount of passengers in the summer months, June to August.
## In the year 1960, July and August had the most passengers, shown by the dark blue color.
```

### **Categorical Plot**

> Using seaborn catplot to visualize categorical data.



```python
## Loading the seaborn 'titanic' dataset to be used in this visualization.
df1 = sns.load_dataset('titanic')
df1.sample(5)
```

```python
# Creating a categorical plot of the 'class' column and graphed by amount of females and males in each class.
sns.catplot(df1, x='class', hue='sex', kind='count', palette='GnBu')

# Using matplotlib.pyplot to provide the title and x,y labels of the graph.
plt.title('Passengers by Sex and Class')
plt.xlabel('Class')
plt.ylabel('Number of Passengers')
plt.show()

## As shown on the graph, there are a higher number of males than females in each of the three classes.
## The 2nd class as the least number of passengers as compared to the 1st and 3rd class, with the 3rd class having the most passengers.
## The 3rd class have the greatest number of males when compared to the other two classes.
```
